```python
import cv2
import numpy as np
import os
import random

colors = {
    'Group1': (0, 0, 0),
    'Group2': (50, 50, 50),
    'Group3': (100, 100, 100),
    'Group4': (0, 34, 67),
    'Group5': (62, 34, 68),
}

data_root = 'data/phison'
num_image = 500

os.makedirs(data_root, exist_ok=True)
```

# good component

```python
atr = 'good'

for k, v in colors.items():
    os.makedirs(os.path.join(data_root, f'{atr} {k}'), exist_ok=True)
    for i in range(num_image):
        # 创建一个白色背景图像
        width, height = 78, 108
        image = np.full((height, width, 3), 255, dtype=np.uint8)

        # 计算矩形的起始点和结束点，使其位于图像中心
        if k == 'Group1':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group2':
            rectangle_width, rectangle_height = 30, 30  # 矩形的宽度和高度
        elif k == 'Group3':
            rectangle_width, rectangle_height = 60, 32  # 矩形的宽度和高度
        elif k == 'Group4':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group5':
            rectangle_width, rectangle_height = 40, 30  # 矩形的宽度和高度   

        start_x = (width - rectangle_width) // 2
        start_y = (height - rectangle_height) // 2
        end_x = start_x + rectangle_width
        end_y = start_y + rectangle_height
        
        # 随机生成偏移量
        offset_x = np.random.randint(-5, 5)
        offset_y = np.random.randint(-5, 5)

        # 对矩形的起始点和结束点进行偏移
        start_x += offset_x
        start_y += offset_y
        end_x += offset_x
        end_y += offset_y

        # 在图像中心绘制黑色矩形
        cv2.rectangle(image, (start_x, start_y), (end_x, end_y), v , thickness=-1)
        
        noise = np.random.randint(-10, 10, image.shape, dtype=np.int8)
        noisy_image = np.clip(image.astype(np.int16) + noise, 0, 255).astype(np.uint8)

        cv2.imwrite(os.path.join(data_root, f'{atr} {k}', "{:05d}.jpg".format(i)), noisy_image)
```

# broke component

```python
atr = 'broke'

for k, v in colors.items():
    os.makedirs(os.path.join(data_root, f'{atr} {k}'), exist_ok=True)
    for i in range(num_image):
        # 创建一个白色背景图像
        width, height = 78, 108
        image = np.full((height, width, 3), 255, dtype=np.uint8)

       # 计算矩形的起始点和结束点，使其位于图像中心
        if k == 'Group1':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group2':
            rectangle_width, rectangle_height = 30, 30  # 矩形的宽度和高度
        elif k == 'Group3':
            rectangle_width, rectangle_height = 60, 32  # 矩形的宽度和高度
        elif k == 'Group4':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group5':
            rectangle_width, rectangle_height = 40, 30  # 矩形的宽度和高度    

        start_x = (width - rectangle_width) // 2
        start_y = (height - rectangle_height) // 2
        end_x = start_x + rectangle_width
        end_y = start_y + rectangle_height
        
        # 随机生成偏移量
        offset_x = np.random.randint(-5, 5)
        offset_y = np.random.randint(-5, 5)

        # 对矩形的起始点和结束点进行偏移
        start_x += offset_x
        start_y += offset_y
        end_x += offset_x
        end_y += offset_y

        # 随机选择一个角
        random_corner = np.random.choice([0, 1])

        # 计算小三角形的坐标
        triangle_size = np.random.randint(10, 15)  # 三角形的大小
        if random_corner == 0:  # 缺失左上角
            triangle_vertices = [(start_x, start_y), (start_x + triangle_size, start_y), (start_x, start_y + triangle_size)]
        elif random_corner == 1:  # 缺失右上角
            triangle_vertices = [(end_x - triangle_size, start_y), (end_x, start_y), (end_x, start_y + triangle_size)]
        elif random_corner == 2:  # 缺失右下角
            triangle_vertices = [(end_x, end_y - triangle_size), (end_x, end_y), (end_x - triangle_size, end_y)]
        elif random_corner == 3:  # 缺失左下角
            triangle_vertices = [(start_x, end_y - triangle_size), (start_x, end_y), (start_x + triangle_size, end_y)]

        # 在图像中心绘制灰色矩形（128, 128, 128）
        cv2.rectangle(image, (start_x, start_y), (end_x, end_y), v, thickness=-1)

        noise = np.random.randint(-10, 10, image.shape, dtype=np.int8)
        noisy_image = np.clip(image.astype(np.int16) + noise, 0, 255).astype(np.uint8)

        # 在图像中绘制缺少的小三角形（黑色）
        cv2.fillPoly(noisy_image, [np.array(triangle_vertices)], (255, 255, 255))
        cv2.imwrite(os.path.join(data_root, f'{atr} {k}', "{:05d}.jpg".format(i)), noisy_image)
```

# shift component

```python
atr = 'shift'

for k, v in colors.items():
    os.makedirs(os.path.join(data_root, f'{atr} {k}'), exist_ok=True)
    for i in range(num_image):
        # 创建一个白色背景图像
        width, height = 78, 108
        image = np.full((height, width, 3), 255, dtype=np.uint8)

        # 计算矩形的起始点和结束点，使其位于图像中心
        if k == 'Group1':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group2':
            rectangle_width, rectangle_height = 30, 30  # 矩形的宽度和高度
        elif k == 'Group3':
            rectangle_width, rectangle_height = 60, 32  # 矩形的宽度和高度
        elif k == 'Group4':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group5':
            rectangle_width, rectangle_height = 40, 30  # 矩形的宽度和高度      
        start_x = (width - rectangle_width) // 2
        start_y = (height - rectangle_height) // 2
        end_x = start_x + rectangle_width
        end_y = start_y + rectangle_height

        # 在图像中心绘制灰色矩形（128, 128, 128）
        cv2.rectangle(image, (start_x, start_y), (end_x, end_y), v, thickness=-1)
    
        # 随机生成小于45度的旋转角度
        angle = np.random.randint(5, 30)

        # 计算旋转矩阵
        center = ((start_x + end_x) // 2, (start_y + end_y) // 2)
        rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1)

        # 应用旋转矩阵到矩形
        rotated_rectangle = cv2.warpAffine(image, rotation_matrix, (width, height), borderMode=cv2.BORDER_CONSTANT, borderValue=(255, 255, 255))
        noise = np.random.randint(-10, 10, rotated_rectangle.shape, dtype=np.int8)
        noisy_image = np.clip(rotated_rectangle.astype(np.int16) + noise, 0, 255).astype(np.uint8)

        cv2.imwrite(os.path.join(data_root, f'{atr} {k}', "{:05d}.jpg".format(i)), noisy_image)
```

```python
atr = 'negative'

for k, v in colors.items():
    os.makedirs(os.path.join(data_root, f'{atr} {k}'), exist_ok=True)
    for i in range(num_image):
        # 创建一个白色背景图像
        width, height = 78, 108
        image = np.full((height, width, 3), 255, dtype=np.uint8)

        # 计算矩形的起始点和结束点，使其位于图像中心
        if k == 'Group1':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group2':
            rectangle_width, rectangle_height = 30, 30  # 矩形的宽度和高度
        elif k == 'Group3':
            rectangle_width, rectangle_height = 60, 32  # 矩形的宽度和高度
        elif k == 'Group4':
            rectangle_width, rectangle_height = 30, 64  # 矩形的宽度和高度
        elif k == 'Group5':
            rectangle_width, rectangle_height = 40, 30  # 矩形的宽度和高度   

        start_x = (width - rectangle_width) // 2
        start_y = (height - rectangle_height) // 2
        end_x = start_x + rectangle_width
        end_y = start_y + rectangle_height
        
        # 随机生成偏移量
        offset_x = np.random.randint(-5, 5)
        offset_y = np.random.randint(-5, 5)

        # 对矩形的起始点和结束点进行偏移
        start_x += offset_x
        start_y += offset_y
        end_x += offset_x
        end_y += offset_y

        # 计算左上角缺失的小三角形坐标
        triangle_size = np.random.randint(10, 15)  # 三角形的大小
        triangle_vertices_left = [(start_x, start_y), (start_x + triangle_size, start_y), (start_x, start_y + triangle_size)]

        # 计算右上角缺失的小三角形坐标
        triangle_vertices_right = [(end_x - triangle_size, start_y), (end_x, start_y), (end_x, start_y + triangle_size)]

        # 在图像中心绘制灰色矩形（128, 128, 128）
        cv2.rectangle(image, (start_x, start_y), (end_x, end_y), v, thickness=-1)

        noise = np.random.randint(-10, 10, image.shape, dtype=np.int8)
        noisy_image = np.clip(image.astype(np.int16) + noise, 0, 255).astype(np.uint8)

        # 在图像中绘制缺少的小三角形（黑色）
        cv2.fillPoly(noisy_image, [np.array(triangle_vertices_left)], (255, 255, 255))
        cv2.fillPoly(noisy_image, [np.array(triangle_vertices_right)], (255, 255, 255))
        cv2.imwrite(os.path.join(data_root, f'{atr} {k}', "{:05d}.jpg".format(i)), noisy_image)
```
